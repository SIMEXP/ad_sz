```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
tau_data = Path("__file__").resolve().parents[1] / 'data' / 'UCBERKELEYAV1451_04_26_22.csv'
```

```python
df['META_ROI'] = df[result_cols].mean(axis=1)
df.head()
```

```python
results_df = df[['EXAMDATE','META_ROI']].copy()
results_df.head()
```
